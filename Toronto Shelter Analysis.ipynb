# Toronto Shelter Analysis

Shelter capacity in the city of Toronto is always a large issue come the winter. News outlets state year over year growth in utilization trends, and commonly share stories about people being turned away. Homelessness is a big issue - especially in the one of the most expensive cities in North America. 

Data maintained by the City of Toronto at https://open.toronto.ca/dataset/daily-shelter-occupancy/

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import glob
```

```python
#Import data from stored dataframes
df = pd.DataFrame()
for filename in glob.glob('Data/daily-shelter-occupancy*.csv'):
    df=df.append(pd.read_csv(filename, parse_dates=True),  sort=False)
    
df.columns = map(str.lower, df.columns)
df.head()
```

## Data Cleaning
Check for and fill NAs

```python
#Check for NAs
df.isna().sum()
```

```python
#Drop id columns
df.drop(['_id','id'], axis=1, inplace=True)
```

```python
#fill capacity NAs using monthly averages

#convert datetime format, then extract month & year
df.occupancy_date = pd.to_datetime(df.occupancy_date,infer_datetime_format=True)

df['year'] = pd.DatetimeIndex(df['occupancy_date']).year
df['month'] = pd.DatetimeIndex(df['occupancy_date']).month
df['week'] = pd.DatetimeIndex(df['occupancy_date']).week
df['quarter'] = pd.DatetimeIndex(df['occupancy_date']).quarter


#find monthly capacity for each shelter - assuming it changes seasonally
df_cap = df.groupby(by=['week','year','shelter_name'], as_index=False)['capacity'].mean()
df_cap.columns = ['week', 'year', 'shelter_name', 'mean_capacity']

#join mean capacity value back to 
df = df.merge(df_cap, on=['week', 'year', 'shelter_name'])

#fill na where exists with the mean capacity
df.capacity = np.where(df.capacity.isnull(), df.mean_capacity, df.capacity)
```

```python
df.isna().sum()
```

```python
df_cap2 = df.groupby(by=['quarter','year','shelter_name'], as_index=False)['capacity'].mean()
df_cap2.columns = ['quarter', 'year', 'shelter_name', 'qmean_capacity']
#join mean capacity value back to 
df = df.merge(df_cap2, on=['quarter', 'year', 'shelter_name'])

#fill na where exists with the mean capacity
df.capacity = np.where(df.capacity.isnull(), df.qmean_capacity, df.capacity)
```

```python
pd.DataFrame(df.groupby(by='shelter_name')['capacity'].mean()) \
    .sort_values(by='capacity',ascending=False) \
    .plot(kind='bar',figsize = (15,10))


plt.show()
```
