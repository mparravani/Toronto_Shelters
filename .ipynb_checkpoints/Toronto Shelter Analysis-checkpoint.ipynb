# Toronto Shelter Analysis

Shelter capacity in the city of Toronto is always a large issue come the winter. News outlets state year over year growth in utilization trends, and commonly share stories about people being turned away. Homelessness is a big issue - especially in the one of the most expensive cities in North America. 

Shelter data maintained by the City of Toronto at https://open.toronto.ca/dataset/daily-shelter-occupancy/   
Weather data extracted for Toronto City Center from Environment Canada at https://climate.weather.gc.ca/historical_data/search_historic_data_e.html


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import glob

# Set standard figure size
from matplotlib import rcParams
rcParams['figure.figsize'] = 15,10
```

```python
#Import data from stored dataframes
df = pd.DataFrame()
for filename in glob.glob('Data/daily-shelter-occupancy*.csv'):
    df=df.append(pd.read_csv(filename, parse_dates=True),  sort=False)
    
df.columns = map(str.lower, df.columns)
df.head()
```

## Data Cleaning
Check for and fill NAs

```python
#Check for NAs
df.isna().sum()
```

```python
#Drop id columns
df.drop(['_id','id'], axis=1, inplace=True)

#NAs in postal code are not important - we are not doing geospatial analysis at this time
df.drop(['shelter_postal_code'], axis=1, inplace=True)

#fill capacity NAs using monthly averages

#convert datetime format, then extract month & year
df.occupancy_date = pd.to_datetime(df.occupancy_date,infer_datetime_format=True)

df['year'] = pd.DatetimeIndex(df['occupancy_date']).year
df['month'] = pd.DatetimeIndex(df['occupancy_date']).month
df['week'] = pd.DatetimeIndex(df['occupancy_date']).week
df['quarter'] = pd.DatetimeIndex(df['occupancy_date']).quarter

#find monthly capacity for each shelter - assuming it changes seasonally
df_cap = df.groupby(by=['week','year','shelter_name'], as_index=False)['capacity'].mean()
df_cap.columns = ['week', 'year', 'shelter_name', 'mean_capacity']

#join mean capacity value back to 
df = df.merge(df_cap, on=['week', 'year', 'shelter_name'])

#fill na where exists with the mean capacity
df.capacity = np.where(df.capacity.isnull(), df.mean_capacity, df.capacity)
```

```python
df.isna().sum()
```

```python
#Attempt to fill remaining NAs with a quarterly average

df_cap2 = df.groupby(by=['quarter','year','shelter_name'], as_index=False)['capacity'].mean()
df_cap2.columns = ['quarter', 'year', 'shelter_name', 'qmean_capacity']

#join mean capacity value back to 
df = df.merge(df_cap2, on=['quarter', 'year', 'shelter_name'])

#fill na where exists with the quarterly mean capacity
df.capacity = np.where(df.capacity.isnull(), df.qmean_capacity, df.capacity)

#drop added capacity columns - not needed
df.drop(columns = ['mean_capacity','qmean_capacity'],inplace=True)
```

```python
#Plot average capacity by shelter
pd.DataFrame(df.groupby(by='shelter_name')['capacity'].mean()) \
    .sort_values(by='capacity',ascending=False) \
    .plot(kind='bar',figsize = (12,8), legend=False)
plt.xlabel('Shelter')
plt.ylabel('Capacity')
plt.title('Average Shelter Capacity')
plt.legend


plt.show()
```

```python
#fill remaining capacity NAs with global average for those shelters, then 0 
df.capacity = df.groupby('shelter_name').transform(lambda x: x.fillna(x.mean()))
df.capacity.fillna(0,inplace=True)
```

## Exploratory Analysis
Questions to be asked;
1 - What are the seasonality trends for utilization and capacity
2 - how does weather affect shelter utilization


### What are the seasonality trends for utilization and capacity

```python
df_daily = pd.DataFrame(df.groupby(by=['occupancy_date','sector'], as_index = False)['occupancy'].sum())

sns.lineplot(data = df_daily, x='occupancy_date', y='occupancy', hue='sector')
plt.xlabel('Date')
plt.ylabel('Occupancy')
plt.title('Shelter Utilization')

plt.show()
```

No strong seasonal trends from the above plot - looks to be more weekly trends, and potentially impacted by capacity available

The growth in family shelter utilization is surprising - I would have assumed growth in single person shelter utilization.

The spiking for the 2017 family data is concerning. It may be due to shelter capacity availability, weekly trends, or poor data capture. 

```python
#Plot utilization as % of capacity
df_daily = pd.DataFrame(df.groupby(by=['occupancy_date','sector'], as_index = False)['occupancy','capacity'].sum())
df_daily['utilization'] = df.occupancy / df.capacity *100

sns.lineplot(data = df_daily, x='occupancy_date', y='utilization', hue='sector')
plt.xlabel('Date')
plt.ylabel('Utilization (%)')
plt.title('Shelter Utilization')

plt.show()
```

When looking at sectors separately, all utilizations seem to jump back and forth between 100% and 50%. What's the distribution of utilizations?

```python
#Distribution of Utilization by sector - Violin Plot
sns.violinplot(y = 'utilization',x='sector',data = df_daily)
plt.xlabel('Sector')
plt.ylabel('Utilization')
plt.title('Shelter Utilization Distribution')

plt.show()


#Distribution of Utilization by sector - Distribution Plot
for sector in set(df_daily.sector):
    sns.distplot(df_daily.utilization[df_daily.sector == sector],hist=False)

plt.xlabel('Utilization')
plt.ylabel('KDE - Utilization')
plt.title('Shelter Utilization Distribution')

plt.show()
```

It looks as though the distributions of utilization are the same regardless of sector. With a high proportion of days at or around 100%
